```python
import pandas as pd

df = pd.read_csv("railroads_data.csv")

df.head(20)
```

```python
df.describe()
```

```python
df.nunique()
```

```python
# Calculates the number of duplicate values and removes them
num_dups = df.duplicated().sum()
print(num_dups)
df = df.drop_duplicates()
num_dups = df.duplicated().sum()
print(num_dups)
```

```python
# Calculates the sum of null values for each column
df.isnull().sum()
```

```python
# Check the percentage of missing values per column
missing_values_percentage = df.isna().mean() * 100

# Check for rows with at least one null value
rows_with_nulls = df[df.isna().any(axis=1)]

# Analyze if there are patterns in missing data
missing_patterns = rows_with_nulls.isna().sum(axis=1).value_counts()

# Prepare a report for the user
missing_report = {
    "Total_Rows": df.shape[0],
    "Total_Columns": df.shape[1],
    "Rows_with_Nulls": rows_with_nulls.shape[0],
    "Rows_without_Nulls": df.dropna().shape[0],
    "Missing_Values_Per_Column": missing_values_percentage,
    "Distribution_of_Missing_Values_in_Rows": missing_patterns
}



missing_report
```

```python
# Drop rows with any number of null values
df = df.dropna()
print(df.shape)
```

```python
import matplotlib.pyplot as plt

# Group by 'Company_Name' to sum the total number of trains per company
trains_per_company = df.groupby('Company_Name')['Number_of_Trains'].sum().sort_values(ascending=False)

# Create a bar chart to compare the total number of trains per company
plt.figure(figsize=(10, 6))
trains_per_company.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Total Number of Trains per Company', fontsize=16)
plt.xlabel('Company Name', fontsize=12)
plt.ylabel('Total Number of Trains', fontsize=12)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Group by 'Company_Name' and calculate the total number of accidents and average safety compliance score
accidents_safety_data = df.groupby('Company_Name').agg(
    Total_Accidents=('Accidents_2023', 'sum'),
    Avg_Safety_Score=('Safety_Compliance_Score', 'mean')
).sort_values('Total_Accidents', ascending=False)

# Create a dual-axis chart to compare accidents and safety compliance score
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar chart for total accidents
color = 'tab:red'
ax1.set_xlabel('Company Name', fontsize=12)
ax1.set_ylabel('Total Accidents (2023)', color=color, fontsize=12)
ax1.bar(accidents_safety_data.index, accidents_safety_data['Total_Accidents'], color=color, alpha=0.7)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xticklabels(accidents_safety_data.index, rotation=45, ha='right')

plt.title('Comparison of Accidents per Company')
fig.tight_layout()

plt.show()
```

```python
# Group by 'Company_Name' and calculate total cargo tonnage and total revenue
cargo_revenue_data = df.groupby('Company_Name').agg(
    Total_Tons=('Cargo_Tonnage_Tons', 'sum'),
    Total_Revenue=('Revenue_USD', 'sum')
).reset_index()

# Create a scatter plot to compare cargo tonnage with revenue
plt.figure(figsize=(10, 6))
plt.scatter(cargo_revenue_data['Total_Tons'], cargo_revenue_data['Total_Revenue'], color='green', alpha=0.7)

# Add titles and labels
plt.title('Comparison of Cargo Tonnage and Revenue per Company', fontsize=16)
plt.xlabel('Total Cargo Tonnage (Tons)', fontsize=12)
plt.ylabel('Total Revenue (USD)', fontsize=12)

# Add labels to points for better understanding
for i, row in cargo_revenue_data.iterrows():
    plt.text(row['Total_Tons'], row['Total_Revenue'], row['Company_Name'], fontsize=9, ha='right')

plt.tight_layout()

plt.show()
```

```python
# Does being on time result in more revenue?

df1 = df.groupby('Company_Name')[['On_Time_Performance_Percent', 'Revenue_USD']].mean()
df1.sort_values(by='Revenue_USD', ascending=False)
```

```python
# Do more accidents result in more maintenance expenses?

df2 = df.groupby('Company_Name')[['Maintenance_Expenses_USD', 'Accidents_2023']].sum()
df2.sort_values(by='Accidents_2023', ascending=False)
```

```python
# How many employees per train per company?

df['Employees_per_Train'] = df['Number_of_Employees'] / df['Number_of_Trains']
df3 = df.groupby('Company_Name')['Employees_per_Train'].mean()
df3.sort_values()
```

```python
#How much does each employee make from each company?

df['Revenue_per_Employee'] = df['Revenue_USD'] / df['Number_of_Employees']
df4 = df.groupby('Company_Name')['Revenue_per_Employee'].mean()
df4.sort_values()
```

```python
# Do older companies generate more revenue than younger companies?

df['Company_Age'] = 2024 - df['Year_Founded']
df5 = df.groupby('Company_Name')[['Company_Age', 'Revenue_USD']].mean()
df5.sort_values(by='Company_Age')
```
